```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
```

```python
def load_dataset(file_path, dataset_name):
    df = pd.read_csv(file_path, sep=' ', header=None)
    df.dropna(axis=1, inplace=True)
    df.columns = ['unit', 'time', 'op1', 'op2', 'op3'] + [f'sensor_{i}' for i in range(1, 22)]
    df['dataset'] = dataset_name
    return df

df_fd002 = load_dataset('/content/train_FD002.txt', 'FD002')
df_fd004 = load_dataset('/content/train_FD004.txt', 'FD004')

# Combine datasets
df = pd.concat([df_fd002, df_fd004], ignore_index=True)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample a small subset for quick plotting
sample_df = df.sample(n=1000, random_state=42)

# Pairplot of operating conditions (to see clusters)
sns.pairplot(sample_df[['op1', 'op2', 'op3']], corner=True)
plt.suptitle("Pairplot of Operating Conditions (op1, op2, op3)", y=1.02)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Scale op1-op3
scaler = StandardScaler()
op_scaled = scaler.fit_transform(df[['op1', 'op2', 'op3']])

# KMeans clustering into 6 op conditions
kmeans = KMeans(n_clusters=6, random_state=42)
df['op_condition'] = kmeans.fit_predict(op_scaled)
df
```

```python
key_sensors = ['sensor_2', 'sensor_4', 'sensor_7', 'sensor_9', 'sensor_11', 'sensor_13', 'sensor_14', 'sensor_16', 'sensor_18', 'sensor_19']

for sensor in key_sensors:
    plt.figure(figsize=(4, 2))
    sns.boxplot(x='op_condition', y=sensor, data=df)
    plt.title(f"{sensor} distribution per Operating Condition Cluster")
    plt.xlabel("Operating Condition")
    plt.ylabel(sensor)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
